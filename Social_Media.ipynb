# Business Case:
Facebook performance metrics ( Social Media Marketing)

The data is related to posts' published during the year of 2017 on the Facebook's page of a renowned cosmetics brand.

This dataset contains 500 of the 790 rows and part of the features

Number of Instances: 500

Number of Attributes: 19

Attribute information:

It includes 7 features known prior to post publication and 12 features for evaluating post impact

Missing Attribute Values: None

# Feature Engineering

```python
data['Page_total_likes'] = data['new'].str.split(';').str[0]
data['Type'] = data['new'].str.split(';').str[1]
data['Category'] = data['new'].str.split(';').str[2]
data['Post_Month'] = data['new'].str.split(';').str[3]
data['Post_weekday'] = data['new'].str.split(';').str[4]
data['Post_Hour'] = data['new'].str.split(';').str[5]
data['Paid'] = data['new'].str.split(';').str[6]
data['Lifetime_post_total_reach'] = data['new'].str.split(';').str[7]
data['Lifetime_post_total_impressions'] = data['new'].str.split(';').str[8]
data['Lifetime_engaged_users'] = data['new'].str.split(';').str[9]
data['Lifetime_post_consumers'] = data['new'].str.split(';').str[10]
data['Lifetime_post_consumptions'] = data['new'].str.split(';').str[11]
data['Lifetime_post_impressions'] = data['new'].str.split(';').str[12]
data['Lifetime_post_reach'] = data['new'].str.split(';').str[13]
data['Lifetime_people'] = data['new'].str.split(';').str[14]
data['Comment'] = data['new'].str.split(';').str[15]
data['Like'] = data['new'].str.split(';').str[16]
data['Share'] = data['new'].str.split(';').str[17]
data['Total_interactions'] = data['new'].str.split(';').str[18]
```

```python
# drop the new column
data = data.drop('new',axis=1)
```

# Converting the categorical values to numerical values

```python
from sklearn.preprocessing import LabelEncoder
```

```python
enc = LabelEncoder()
```

```python
data.Type = enc.fit_transform(data.Type)
```

```python
data['Page_total_likes'] = data['Page_total_likes'].astype(int)
data['Type'] = data['Type'].astype(int)
data['Category'] = data['Category'].astype(int)
data['Post_Month'] = data['Post_Month'].astype(int)
data['Post_weekday'] = data['Post_weekday'].astype(int)
data['Post_Hour'] = data['Post_Hour'].astype(int)
data['Lifetime_post_total_reach'] = data['Lifetime_post_total_reach'].astype(int)
data['Lifetime_post_total_impressions'] = data['Lifetime_post_total_impressions'].astype(int)
data['Lifetime_engaged_users'] = data['Lifetime_engaged_users'].astype(int)
data['Lifetime_post_consumers'] = data['Lifetime_post_consumers'].astype(int)
data['Lifetime_post_consumptions'] = data['Lifetime_post_consumptions'].astype(int)
data['Lifetime_post_impressions'] = data['Lifetime_post_impressions'].astype(int)
data['Lifetime_post_reach'] = data['Lifetime_post_reach'].astype(int)
data['Lifetime_people'] = data['Lifetime_people'].astype(int)
data['Comment'] = data['Comment'].astype(int)
data['Total_interactions'] = data['Total_interactions'].astype(int)
```
