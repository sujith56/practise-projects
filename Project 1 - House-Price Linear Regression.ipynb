 sqft_base: Sqaure ft in the basement

 yr_renovation: Date of renovation of the house

#### Imorting Libraries

```python
# importing numpy and pandas
import numpy as np
import pandas as pd
```

```python
#imoporting csv data
data = pd.read_csv('C:\\Users\\abhishek\\Desktop\\Machine Learning practice\\House-price-master\\houseprice.csv')
```

```python
# Viewing the first 5 rows of the data
data.head()
```

```python
#checking the shape of the data
data.shape
```

```python
# slicing unsued columns
data = data.iloc[:,2:]
```

```python
data.head()
```

```python
# assigning (y)dependent and (X)independent values
y = data.price
X = data.iloc[:,1:]
```

```python
X.head()
```

```python
# traing and testing
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=251)
```

```python
# importing linear regression model
from sklearn.linear_model import LinearRegression
```

```python
# assinging model with LinearRegression
model = LinearRegression()
```

```python
# fitting model between the X train and y train
model.fit(X_train,y_train)
```

```python
# predicting X_test and y_test
predict_lr = model.predict(X_test)
```

```python
y_test.head(10)
```

```python
predict_lr
```

```python
# Importing r2_score from metrics
from sklearn.metrics import r2_score
```

```python
#Regression score
r2_score(y_test,predict_lr)
```
